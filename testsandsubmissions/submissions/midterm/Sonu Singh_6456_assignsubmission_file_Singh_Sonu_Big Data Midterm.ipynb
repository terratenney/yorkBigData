**48) Briefly describe the differences between "supervised" and "unsupervised" learning.**

```python
Supervised learning: a form of machine learning where a training set is given as input and each input is labeled
    with a desired outcome. This method maps the input data to output of that data. Algorithm is trained iteratively
    until it achieves an acceptable level of performance.
Unsupervised learning: this form of machine learning where a data set is given as input and no desired 
    output is specified. The goal is to model the input data to learn more about the data. This type of learning is 
    more complex
```
