```python
# Volume 
# Volume is a term use to describe Big Data from a relative perspective to information processing
# historically in the sense that things for Big Data the sheer amount of information
# is exponentially in terms size of the information being stored that will be ready to be
# analyzed and worked with.  Big Data needs to use this 'V' description as since volume
# is an important factor as it can impact what technology can be used as it data amounts increase.
# 


# Variety 
# Is a term used to describe Big Data as comprising of many different (non traditional) sources 
# the data could be messages, pictures/images (from media/social networks), data from sensor devices
# (IoT), activity and location data coming from devices/mobile phones all comprising together of 
# information/data which may not be suitable for storing in a structured database form.
# This 'V' is important in playing a role in describing Big Data as when there is more 
# variety newer technologies must be incorporated into situation to keep up with new forms/sources
# of data.


# Velocity
# Is used to describe Big Data as data that is important due to the fact
# that for certain applications, the speed of data creation is more important
# than the volume in terms of solving the need or problem being looked at 
# with the data.  This is important in cases where finding a quick way to information 
# to model a trend can fulfill (or allow for a competitive advantage to be had)
# in answering a question or solving a problem.  This importance of this 'V' has to be considered
# to see how far we can take things (or if there is even a need to do so given the what we are
# trying to answer) as constraints of technology and maybe even the other 'V's can constrain too.

# Veracity 
# Can be used to describe Big Data in terms of how strong our trust is in the data we have
# as with all data there is always an element of error, in-accurracy, uncertainty and mistakes/faults
# in it.
# This 'V' has to be factored in to the situation and used as a way to describe Big Data
# as the level of this 'V' will fluctuate with volume changes in the data, capture techniques/sources,
# and processing the 'noise' or confidence level has to taken into 
# consideration as we use the data (or look at the results from it).

# Value
# This 'V' is used to describe the usefullnes of the Big Data and the benefit and positive impact 
# to solve an issue or answer a question.
# 

```


**10) What is the name for the 3 primary measures of central tendency and give their definition** 

```python
# The 3 primary measures of central tendency are: Mean, Mode, and Median 
# for a data set of numbers.
#
# The following are their definitions.
#
# Mean:
# Is the average of all of the numbers in the set of data.  This is essentially 
# the result of the sum of all numbers in the data set divided by the number 
# (count) of data points in the set.
#
# Mode:
# Is the number (or set of numbers) that occur the most frequent 
# in the data set.  This could be one number if only one occurs more frequent
# than all others in a data set and is above all others in terms of count of
# occurrence.  
#
# There could also be multiple numbers as modes(ie 2 or more)
# if several numbers in the data set are all tied together as
# with the most frequency in terms of occurrence/count in a data set
# 
# Median:  
# Is the middle value of the set, when the numbers in a set are ordered
# in ascending or descending order.  If a set has an odd count of numbers
# the 'Median' is the middle number where the count of numbers both to 
# the left of it and to the right of contain an equal count of numbers.
# When the count of numbers in a set is even.  The 'Median' would be the average
# of the two numbers in the middle positions of the list.
# 
```

**48) Briefly describe the differences between "supervised" and "unsupervised" learning.**

```python
# Supervised learning is where the system is given example/input data which are 
# 'labeled' as correct results and whatever is undesired is 'labeled'
# as incorrect.  Based on the machine analyzing the characteristics of the good and bad examples it will 
# hopefully build a model (learn) to see what reasons (characteristics) make up 
# a desired (good) result so that when presented with a new (unencountered) set of data to analyze, that it can
# apply the model to predict correctly whether or not it has the desired characteristics and through that label it
# properly.
#
# It is called 'Supervised' as there is guidance and assistance
# for teaching the system to learn (build) the algorithm that would generate the correct
# answers may have through the fact that prior labeled input/data us being fed into the system.

# Unsupervised learning is where data (examples) are given to the system, but this time they
# are not labeled as being 'good' or 'bad' ('desired'/'undesired'), the machine
# learning under this method is left to take the data, identify patterns/characterstics (under self created labels), and then
# match similar data together into many groups (labels) based on similarities it has seen or experienced in the past. 
# 
# The hope is that with this type of learning, the system has enough examples ('learned') so that it can apply a label (it self-created)
# correctly for new unencountered data. 
```

**EXTRA CREDIT 51) Why is Ordinary Least Squares Regression, called Ordinary Least Squares?**

```python
# "Ordinary Least Squares Regression" is called Ordinary Least Squares because it is essentially trying to
# establish a linear relationship between one variable/factor by trying the find the line that fits within observed data points
# that yields the minimal (least) sum of the squares distance between the actual observed data/results and the line we are 
# fitting into the actual data points.  When there is only one factor (or variable) the situation 
# statistically/mathematically speaking is said to be 'Ordinary' as the function (in this case a line function) has a 
# single variable.  To contrast this, if we had a fuction/situation where there were multiple variables/factors it would not
# be called 'Ordinary Least Squares', we have to use partial least squares regression as a method to establish 
# a correlation if there is more than one variable.
```

**EXTRA CREDIT 53) What are WMD's according to Cathy O'Neil?**

```python
# WMD is an acronym for "Weapons of Math Destruction".  
# A WMD according to Cathy O'Neil is:
#
#  A mathematical model that is used on a large scale that was supposed to do something good, but as it was used and applied
#  turned out harmful with unintended consequences that ultimately self creates more data that reinforces the model and shows its
#  success.
#
# WMDs usually have the a lot of the following characteristics:
#    -They impact on a discriminatory manner (ie harming the poor).
#    -How they work is not transparent, things may be a black box in the sense it's not known (by it's subjects):
#        -how information is being collected
#        -what data/information is being used 
#        -no way to question or get answers about reasons for results being the way they are
#        -models results can be cheated on (or 'gamed') by people which can result in situations where the opposite 
#         to what was intended for the model to achieve happens. 
#    -There is no feedback/improvement mechanism for models that are considered WMDs, there is no circle back 
#     to see if improvement is needed (in the cases where things went wrong) and why the results that actually happened
#     differed from those predicted by the model, there is no review done to see if elements of the model need to 
#     be changed, tweaked, redesigned, whether or not the data is correct and even if there are incorrect assumptions 
#     in the picture.
# 
```

**EXTRA CREDIT 54) Why do vicious feedback loops matter in data analytics?**

```python
# Vicious feedback loops matter in data analytics as it can cause a runaway situation where the predicted results 
# from the model can generate data that further reinforces the success of the model and hides
# incorrect assumptions that went into the building of the model itself and ultimately obscures the fact that the
# model has a flaw, either in the data being used as input, or the algorithm itself (used to generate the result)
# has issues.
#
# When such situations happen and a model is used on a large scale there can be a negative impact  
# as the model (or data it generates) increasingly happens to more and more people or causes impact to focus in a more
# narrower and narrower segment of the world/population in an uneven and unfair manner.
```
