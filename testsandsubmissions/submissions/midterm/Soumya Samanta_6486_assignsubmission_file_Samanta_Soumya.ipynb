**48) Briefly describe the differences between "supervised" and "unsupervised" learning.**

```python
Supervised learning is when we build a model based on input and desired output values. Labelled training data is available 
and the training data consists of training examples.

Unsupervised learning is when we have unlabelled data and hidden functions are inferred to describe hidden structures. 
Imagine we have unlabelled data and we are trying to group together similar data points based off of features.

Example, say we are trying to create a prediction model to say of an input is Male or Female, based length of hair. 
Supervised Learning - If we have a set of labelled data [Sex, LengthOfHair], we can train the model. Now say when we input the 
length of hair, the model will be able to 'predict' the sex of person.
Unsupervised Learning - We do not have set of labelled data. The only input to the model is all the [lengthofhair]. From this,
the model may be able to group together the data points.
```
