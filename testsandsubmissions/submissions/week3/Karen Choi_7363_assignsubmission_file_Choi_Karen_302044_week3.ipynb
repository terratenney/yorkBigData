```python
f = np.random.normal(0, 1, size=30)
```

```python
bandwidth = 1.08 * f.std()* f.size ** (-1/5.)
```

```python
support = np.linspace(-4,4,200)
```

```python
kdeg = []
for f_i in f:

    kde = stats.norm(f_i, bandwidth).pdf(support)
    kdeg.append(kde)
    plt.plot(support, kde, color="b")
```

```python
density = np.sum(kdeg, axis=0)
density /= integrate.trapz(density, support)
plt.plot(support, density);
```
