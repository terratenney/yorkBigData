```python
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as ply
```

```python
import seaborn as sns
sns.set(color_codes=True)
```

```python
np.random.seed(sum(map(ord, "distributions")))
```

```python
r = np.random.normal(size=100)
sns.distplot(r);
```

```python
sns.distplot(r, kde=False, rug=True);
```

```python
sns.distplot(r, bins=20, kde=False, rug=True);
```

```python
sns.distplot(r, hist=False, rug=True);
```

```python
x = np.random.normal(0, 1, size=30)
bandwidth = 1.06 * r.std() * r.size **(-1/5.)
support = np.linspace(-4, 4, 200)

kernels = []
for r_i in r:
    
    kernel = stats.norm(r_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color="g")
    
sns.rugplot(r, color=".2", linewidth=3);
```

```python
dens = np.sum(kernels, axis=0)
dens /= integrate.trapz(dens, support)
ply.plot(support, dens);
```
