```python
# convert ints / strings to floats, give up on anything else (call it 0.0)
def floatify(val):
    if val == None or val == 'None':
        return 0.0
    else:
        return float(val)
```

```python
#this will render 5 curves
x_5 = np.random.normal(0, 1, size=5)
bandwidth = 1.06 * x_5.std() * x_5.size ** (-1 / 5.)
support = np.linspace(-4, 4, 200)

kernels_5 = []
for x_i in x_5:

    kernel = stats.norm(x_i, bandwidth).pdf(support)
    kernels_5.append(kernel)
    plt.plot(support, kernel, color="r")

sns.rugplot(x_5, color=".2", linewidth=3);
```

```python
x = np.random.normal(0, 1, size=30)
bandwidth = 1.06 * x.std() * x.size ** (-1 / 5.)
support = np.linspace(-4, 4, 200)

kernels = []
for x_i in x:

    kernel = stats.norm(x_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color="r")

sns.rugplot(x, color=".2", linewidth=3);
```

```python
#test using 5 curves
density = np.sum(kernels_5, axis=0)
density /= integrate.trapz(density, support)
plt.plot(support, density);
```

```python
density = np.sum(kernels, axis=0)
density /= integrate.trapz(density, support)
plt.plot(support, density);
```
