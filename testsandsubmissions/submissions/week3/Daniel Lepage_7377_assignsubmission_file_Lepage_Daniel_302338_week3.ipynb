```python
dan = np.random.normal(0,1, size = 50)
bandwidth = 1 * dan.std() * dan.size ** (-1/5.)
newsupport1 = np.linspace(-4, 4, 400)
newkernels1 = []
for dan_i in dan:

    newkernel1 = stats.norm(dan_i, bandwidth).pdf(newsupport1)
    newkernels1.append(newkernel1)
    plt.plot(newsupport1, newkernel1, color="y")
    
sns.rugplot(dan, color=".5", linewidth = 5)
    
```

```python
newdensity1 = np.sum(newkernels1, axis = 0)
newdensity1 /= integrate.trapz(newdensity1, newsupport1)
plt.plot(newsupport1, newdensity1);
```
