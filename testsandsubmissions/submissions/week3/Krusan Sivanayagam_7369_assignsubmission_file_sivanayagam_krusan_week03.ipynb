```python
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plot
```

```python
import seaborn as sb
sb.set(color_codes=True)
```

```python
np.random.seed(sum(map(ord, "distributions")))
```

```python
a = np.random.normal(size=100)
sb.distplot(a);
```

```python
sb.distplot(a, kde=False, rug=True);
```

```python
sb.distplot(a, bins=20, kde=False, rug=True);
```

```python
sb.distplot(a, hist=False, rug=True);
```

```python
a = np.random.normal(0, 1, size=30)
bandwidth = 1.06 * a.std() * a.size ** (-1 / 5.)
support = np.linspace(-4, 4, 200)

kernels = []
for a_i in a:

    kernel = stats.norm(a_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color="r")

sb.rugplot(a, color=".2", linewidth=3);
```

```python
dense = np.sum(kernels, axis=0)
dense /= integrate.trapz(dense, support)
plot.plot(support, dense);
```

```python
sb.kdeplot(a, shade=True);
```

```python
sb.kdeplot(a)
sb.kdeplot(a, bw=.2, label="bw: 0.2")
sb.kdeplot(a, bw=2, label="bw: 2")
plot.legend();
```

```python
sb.kdeplot(a, shade=True, cut=0)
sb.rugplot(a);
```
