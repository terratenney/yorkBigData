```python
dataf1[dataf1.twitter_lang == 'pt'].head()
```

```python
dataf2 = dataf1[["actor:utcOffset","actor:followersCount", "actor:friendsCount", "actor:listedCount"]]

dataf2.head()
```

```python
def floatify(x):
    if x == None or x == 'None':
        return 0.0
    else:
        return float(x)
```

# Visualizing the distribution of a dataset

```python
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set(color_codes=True)
```

```python
np.random.seed(sum(map(ord, "distributions")))
```

```python
dist = np.random.normal(size=200)
sns.distplot(dist);
```

**Histograms**


```python
sns.distplot(dist, kde=False, rug=True);
```

```python
sns.distplot(dist, bins=30, kde=False, rug=True);
```

**Kernel density estimaton**


```python
sns.distplot(dist, hist=False, rug=True);
```

```python
dist = np.random.normal(0, 1, size=30)
bandwidth = 1.06 * dist.std() * dist.size ** (-1 / 5.)
support = np.linspace(-5, 5, 200)

kernels = []
for dist_i in dist:

    kernel = stats.norm(dist_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color="b")

sns.rugplot(dist, color=".2", linewidth=3);
```

```python
D = np.sum(kernels, axis=0)
D /= integrate.trapz(D, support)
plt.plot(support, D);
```

```python
sns.kdeplot(dist, shade=True);
```

```python
sns.kdeplot(dist)
sns.kdeplot(dist, bw=.1, label="bw: 0.1")
sns.kdeplot(dist, bw=1, label="bw: 1")
sns.kdeplot(dist, bw=2, label="bw: 2")
sns.kdeplot(dist, bw=.5, label="bw: 0.5")
plt.legend();
```

```python
sns.kdeplot(dist, shade=True, cut=2)
sns.rugplot(dist);
```

**Fitting parametric distributions**


```python
data = np.random.gamma(6, size=200)
sns.distplot(data, kde=False, fit=stats.gamma);
```

Plotting bivariate distributions
--------------------------------


```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
dat = np.random.multivariate_normal(mean, cov, 200)
datfr = pd.DataFrame(dat, columns=["x", "y"])
```

**Scatterplots**

```python
sns.jointplot(x="x", y="y", data=datfr);
```

**Hexbin plots**


```python
A, B = np.random.multivariate_normal(mean, cov, 2000).T
with sns.axes_style("white"):
    sns.jointplot(x=A, y=B, kind="hex", color="k");
```

**Kernel density estimation**


```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
dat = np.random.multivariate_normal(mean, cov, 3000)
datfr = pd.DataFrame(dat, columns=["A", "B"])
sns.jointplot(x="A", y="B", data=datfr, kind="kde");
```

```python
f, ax = plt.subplots(figsize=(6, 6))
sns.kdeplot(datfr.A, datfr.B, ax=ax)
sns.rugplot(datfr.A, color="g", ax=ax)
sns.rugplot(datfr.B, vertical=True, ax=ax);
```

```python
f, ax = plt.subplots(figsize=(6, 6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(datfr.A, datfr.B, cmap=cmap, n_levels=60, shade=True);
```

```python
g = sns.jointplot(x="A", y="B", data=datfr, kind="kde", color="g")
g.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$A$", "$B$");
```
