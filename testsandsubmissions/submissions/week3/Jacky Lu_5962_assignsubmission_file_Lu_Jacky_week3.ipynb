```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
Fdata = np.random.multivariate_normal(mean, cov, 100)
df1 = pd.DataFrame(Fdata, columns=["x", "y"])
```

```python
x.jointplot(x="x", y="y", data=df1);
```
