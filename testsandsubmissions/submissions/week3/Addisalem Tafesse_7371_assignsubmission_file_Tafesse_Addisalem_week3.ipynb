```python
p = np.random.normal(0, 1, size=20)
bandwidth = 1.05 * p.std() * p.size **(-1/5.)
support = np.linspace(-5, 5, 200)
kernels = []
for p_i in p:
    kernel = stats.norm(p_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color="g")

sns.rugplot(p, color=".15", linewidth=3);
```

```python
dnsty = np.sum(kernels, axis=0)
dnsty /= integrate.trapz(dnsty, support)
plt.plot(support, dnsty);
```
