# Slicing and Combining

```python
df1[df1.twitter_lang == 'en'].head()
```

```python
df2 = df1[["gnip:klout_score","actor:followersCount", "actor:friendsCount", "actor:listedCount"]]
df2.head(10)
```

```python
df2.dtypes
```

```python
def floatify(val):
    if val == None or val == 'None':
        return 0.0
    else:
        return float(val)
```

```python
df2['gnip:klout_score'] = df2['gnip:klout_score'].map(floatify)
df2.dtypes
```

```python
df2 = df2.astype(float)
df2.dtypes
```

```python
df2['total_score/count'] = df2['gnip:klout_score'] / df2['actor:followersCount']
df2.head(10)
```

# Grouping

```python
df1.head(10)
```

```python
pop_df = df1[df1["actor:followersCount"] >=100]
```

```python
lang_gb = pop_df[['twitter_lang',\
                  'gnip:klout_score',\
                  'actor:followersCount',\
                  'actor:friendsCount',\
                  'actor:statusesCount']].groupby('twitter_lang')
lang_gb.head()
```

```python
lang_gb_mean = lang_gb.aggregate(np.mean)
lang_gb_mean.head()
```

# Plotting univariate distributions

```python
x = np.random.normal(size=100)
sns.distplot(x);
```

```python
sns.distplot(x, kde=False, rug=True);
```

```python
sns.distplot(x, hist=False, rug=True);
```

# Plotting bivariate distributions

```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])
```

```python
sns.jointplot(x="x", y="y", data=df);
```

```python
x, y = np.random.multivariate_normal(mean, cov, 1000). T
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="hex", color="k");
```

```python
sns.jointplot(x="x", y="y", data=df, kind="kde");
```

# Visualizing pairwise relationships in a dataset

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris);
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, cmap="Blues_d", n_levels=6);
```
