**19) What does the statistic Standard Deviation tell you about a numerical dataset? **

```python
print "Standard Deviation (SD) is a measure of variance within a data set, calculated based on the mean. A low SD is indicative of data points that vary within a smaller range, i.e are closer to the mean of the data set. Conversely, a higher SD indicates that the data points are spread over a wider range of values, i.e further away from the set mean. For a given data set, the SD is the square root of the variance. To calculate the SD, the differene between each data point and the mean is calculated and squared; the squared differences are then summed, and the sum is divided by the number of data points -1. The resultant value is then square rooted."
```
