```python
# In[59]  convert ints / strings to floats, give up on anything else (call it 0.0)
def floatify(value):
    if value == None or value == 'None':
        return 0.0
    else:
        return float(value)
```

```python
s1 = np.random.normal(0, 1, size=30)
bandwidth = 1.06 * s1.std() * s1.size ** (-1 / 5.)
support = np.linspace(-4, 4, 200)

kernels = []
for s1_i in s1:

    kernel = stats.norm(s1_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color="y")

sns.rugplot(s1, color=".2", linewidth=3);
```

```python
density = np.sum(kernels, axis=0)
density /= integrate.trapz(density, support)
plt.plot(support, density);
```
